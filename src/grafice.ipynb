```python
import numpy as np
import matplotlib.pyplot as plt

def superClean(a):
    a = (a + 6) / 4
    c = np.zeros(22050)
    for d in range(22050):
        e = 2 * d / 22050 - 1
        c[d] = (1 + a) * e / (1 + a * abs(e))
    return np.linspace(-1, 1, 22050), c

def crunch(a):
    a = a**2
    c = np.zeros(22050)
    for d in range(22050):
        f = 2 * d / 22050 - 1
        c[d] = (1 + a) * f / (1 + a * abs(f))
    return np.linspace(-1, 1, 22050), c

# Set the saturation level
a = 3

# Generate curves for clean channel
x_superClean, y_superClean = superClean(a)
x_crunch, y_crunch = crunch(a)

# Plotting side-by-side
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(x_superClean, y_superClean, color='green')
axs[0].set_title('Curba pentru primul nod')
axs[0].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axs[0].grid(True)

axs[1].plot(x_crunch, y_crunch, color='green')
axs[1].set_title('Curba pentru al doilea nod')
axs[1].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def classic_distortion(k, n_samples=22050):
    deg = np.pi / 180
    x = np.linspace(-1, 1, n_samples)
    curve = (3 + k) * x * 57 * deg / (np.pi + k * np.abs(x))
    return x, curve

def high_distorted(a, samples=22050):
    a = (a + 2) ** 3
    x = np.linspace(-1, 1, samples)
    curve = (1 + a) * x / (1 + a * np.abs(x))
    return x, curve

# Set the saturation level
a = 3

# Generate curves for distortion channel
x_classic, y_classic = classic_distortion(a)
x_high, y_high = high_distorted(a)

# Plotting side-by-side
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(x_classic, y_classic, color='red')
axs[0].set_title('Curba pentru primul nod (distors)')
axs[0].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axs[0].grid(True)

axs[1].plot(x_high, y_high, color='red')
axs[1].set_title('Curba pentru al doilea nod (distors)')
axs[1].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
